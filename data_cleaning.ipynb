```python
import pandas as pd
import numpy as np

# Loading the dataset
df = pd.read_csv('/content/drive/MyDrive/DSML_Projects/flipkart_sales/flipkart_sales.csv')
```

```python
df.head()
```

```python
# Checking data types and missing values
print(df.info())
print(df.isnull().sum())
```

```python
# Correcting data types
df['Order ID']= df['Order ID'].astype(str)
df['Order Date'] = pd.to_datetime(df['Order Date'],format='%d-%m-%Y')
```

```python
# Validating data (check for negative quantities or prices)
df = df[df['Quantity Sold'] >= 0]
df = df[df['Price (INR)'] >= 0]
df = df[df['Customer Rating'] >= 0]
```

```python
#Standardizing Tiles (removing spaces)
df.columns = df.columns.str.replace(' ', '')

df.info()
```

```python
# Saving the cleaned data
df.to_csv('/content/drive/MyDrive/DSML_Projects/flipkart_sales/flipkart_sales_cleaned.csv', index=False)
```
